````python
import utils

import pandas as pd
import altair as alt
````

* Confuturo data

````python
results_df, corporativo_df, tecnologico_df = utils.read_confuturo_dataframes()
corporativo_df['timestamp'] = corporativo_df['timestamp'].apply(lambda t: t.strftime('%Y-%m-%d'))
tecnologico_df['timestamp'] = tecnologico_df['timestamp'].apply(lambda t: t.strftime('%Y-%m-%d'))
````

````python
from_date = corporativo_df['timestamp'].min()
to_date = corporativo_df['timestamp'].max()
````

````python
# portfolio 1
santander_esg = utils.get_daily_assets(11401, from_date, to_date).rename(columns={'date': 'timestamp'}).set_index('timestamp')
itau = utils.get_daily_assets(10235, from_date, to_date).rename(columns={'date': 'timestamp'}).set_index('timestamp')
larrain = utils.get_daily_assets(10603, from_date, to_date).rename(columns={'date': 'timestamp'}).set_index('timestamp')

# portfolio 2
santander_usa = utils.get_daily_assets(11453, from_date, to_date).rename(columns={'date': 'timestamp'}).set_index('timestamp')
````

````python
moderate_metlife = (0.2*santander_esg + 0.2*itau + 0.5*larrain)/0.9
moderate_metlife = moderate_metlife/moderate_metlife.values[0] * 100
````

* add administrator column

````python
moderate_metlife['origin'] = 'metlife'
santander_usa['origin'] = 'metlife'
corporativo_df['origin'] = 'confuturo'
tecnologico_df['origin'] = 'confuturo'

all_dfs = {
    'moderate_metlife': moderate_metlife.reset_index(),
    'santander_usa': santander_usa.reset_index(),
    'corporativo_df': corporativo_df,
    'tecnologico_df': tecnologico_df
}
````

````python
results_df = pd.concat(
    list(all_dfs.values()),
    keys=list(all_dfs.keys()),
    names=["modalidad/fondo", "Row ID"],
)
results_df = results_df.reset_index(level=0)
````

````python
alt.Chart(
    results_df.reset_index(level=0),
    width=600, height=400
).mark_line().encode(
    alt.X("timestamp", type="temporal", axis=alt.Axis(format="%y.%m.%d")),
    alt.Y("ValorBase100", type="quantitative", scale=alt.Scale(zero=False)),
    tooltip=["modalidad/fondo", "timestamp", "ValorBase100"],
#     shape='origin',
    color=alt.Color(field='modalidad/fondo', 
                    type='nominal',
                    scale=alt.Scale(
                        domain=['moderate_metlife', 'santander_usa', 'corporativo_df', 'tecnologico_df'], 
                        range=['orange', 'orange', 'lightblue', 'lightblue']))
).interactive()
````
