````python
import utils

import pandas as pd
import altair as alt
````

* Confuturo data

````python
results_df, corporativo_df, tecnologico_df = utils.read_confuturo_dataframes()
corporativo_df['timestamp'] = corporativo_df['timestamp'].apply(lambda t: t.strftime('%Y-%m-%d'))
tecnologico_df['timestamp'] = tecnologico_df['timestamp'].apply(lambda t: t.strftime('%Y-%m-%d'))
````

````python
from_date = corporativo_df['timestamp'].min()
to_date = corporativo_df['timestamp'].max()
````

* Metlife data

```
10603 FONDO MUTUO LARRAINVIAL CUENTA ACTIVA MODERADA
10235 FONDO MUTUO ITAU DINAMICO
11401 FONDO MUTUO SANTANDER GO ACCIONES GLOBALES ESG
11453 FONDO MUTUO SANTANDER GO ACCIONES USA```

````python
santander_esg = utils.get_daily_assets(11401, from_date, to_date)
santander_usa = utils.get_daily_assets(11453, from_date, to_date)
itau = utils.get_daily_assets(10235, from_date, to_date)
larrain = utils.get_daily_assets(10603, from_date, to_date)
````

````python
dfs = {
    'santander_esg': santander_esg,
    'santander_usa': santander_usa,
    'itau': itau,
    'larrain': larrain
}
for df_name in dfs:
    dfs[df_name] = dfs[df_name].rename(columns={'date': 'timestamp'})
````

````python
results_df = pd.concat(
    [corporativo_df, tecnologico_df] + list(dfs.values()),
    keys=["corporativo", "tecnologico"] + list(dfs.keys()),
    names=["modalidad/fondo", "Row ID"],
)


results_df.reset_index(level=0)
````

````python
alt.Chart(
    results_df.reset_index(level=0),
    width=600, height=400
).mark_line().encode(
    alt.X("timestamp", type="temporal", axis=alt.Axis(format="%y.%m.%d")),
    alt.Y("ValorBase100", type="quantitative", scale=alt.Scale(zero=False)),
    tooltip=["modalidad/fondo", "timestamp", "ValorBase100"],
    color='modalidad/fondo',
).interactive()
````
