```python
itaus = [asset['attributes']['name'] for asset in conceptual_assets['data'] 
 if 'ITAU' in asset['attributes']['name'] 
 and 'DINAMICO' in asset['attributes']['name']
]
itaus
```

* Santander list

```python
santanders = [asset['attributes']['name'] for asset in conceptual_assets['data'] 
 if 'SANTANDER' in asset['attributes']['name'] 
 and (
     'ESG' in asset['attributes']['name']
    or 'USA' in asset['attributes']['name']
 )
]
santanders
```

# Selected names

```python
names = [
    'FONDO MUTUO LARRAINVIAL CUENTA ACTIVA MODERADA',
    'FONDO MUTUO ITAU DINAMICO',
    'FONDO MUTUO SANTANDER GO ACCIONES GLOBALES ESG',
    'FONDO MUTUO SANTANDER GO ACCIONES USA'
]
```

# Get real assets for each name

Show series names

```python
name = names[0]
conceptual_assets_dict = {}
for name in names:
    conceptual_asset = [asset for asset in conceptual_assets['data'] if asset['attributes']['name'] == name][0]
    

    real_asset_request = requests.get(f'https://fintual.cl/api/conceptual_assets/{conceptual_asset["id"]}/real_assets')
    real_asset = real_asset_request.json()
    conceptual_assets_dict[name] = real_asset

    print('\n')
    print(name)
    print([s['attributes']['serie'] for s in real_asset['data']])
    print(100*'-')
```

Pick only those that say "APV"

```python
real_assets_dict = {}
for c_asset in conceptual_assets_dict:
    real_assets_dict[c_asset] = [r_a for r_a in conceptual_assets_dict[c_asset]['data'] if r_a['attributes']['serie'] == 'APV'][0]
```

```python
real_assets_dict
```

# Show ids

```python
for real_asset in real_assets_dict:
    print(real_assets_dict[real_asset]['id'], real_asset)
```
