```python
import datetime
```

```python
def make_df(json_data):
    df = pd.DataFrame(json_data)
    df["timestamp"] = df["FechaUltima"].apply(
        lambda a: datetime.datetime.fromtimestamp(int(float(
            a.replace("/", "").replace("Date", "").replace("(", "").replace(")", "")
        )/1000))
    )
    return df[["timestamp", "ValorBase100"]]


corporativo_df = make_df(corporativo_data["Rentabilidades"])
tecnologico_df = make_df(tecnologico_data["Rentabilidades"])
```

```python
# result = pd.concat([corporativo_df, tecnologico_df], axis=1, sort=False)
results_df = pd.concat([corporativo_df, tecnologico_df], keys=['corporativo', 'tecnologico'], names=['modalidad', 'Row ID'])
results_df
```

```python
results_df.reset_index(level=0)
```

```python
import altair as alt

alt.Chart(
    results_df.reset_index(level=0)
).mark_line().encode(
    alt.X("timestamp", type="temporal", axis=alt.Axis(format="%y.%m.%d")),
    alt.Y("ValorBase100", type="quantitative"),
    tooltip=["timestamp", "ValorBase100"],
    color='modalidad',
).interactive()
```
